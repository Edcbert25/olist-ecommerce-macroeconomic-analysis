```python
import pandas as pd

# 1. Carregar as tabelas principais
orders = pd.read_csv('../dados/dados_brutos_olist/olist_orders_dataset.csv')
items = pd.read_csv('../dados/dados_brutos_olist/olist_order_items_dataset.csv')
products = pd.read_csv('../dados/dados_brutos_olist/olist_products_dataset.csv')
customers = pd.read_csv('../dados/dados_brutos_olist/olist_customers_dataset.csv')
payments = pd.read_csv('../dados/dados_brutos_olist/olist_order_payments_dataset.csv')
reviews = pd.read_csv('../dados/dados_brutos_olist/olist_order_reviews_dataset.csv')

# 2. Converter colunas de data 
cols_data = [
    'order_purchase_timestamp', 
    'order_approved_at', 
    'order_delivered_carrier_date', 
    'order_delivered_customer_date', 
    'order_estimated_delivery_date'
]
for col in cols_data:
    orders[col] = pd.to_datetime(orders[col])

# 3. Cruzamento de Tabelas (O Grande Merge)
# Começamos pelos pedidos e unimos com itens e clientes
master = orders.merge(items, on='order_id', how='left')
master = master.merge(customers, on='customer_id', how='left')
master = master.merge(products, on='product_id', how='left')
master = master.merge(payments, on='order_id', how='left')
master = master.merge(reviews, on='order_id', how='left')

# 4. Criar colunas básicas de tempo para facilitar a vida do grupo
master['ano_mes'] = master['order_purchase_timestamp'].dt.to_period('M')
master['dia_semana'] = master['order_purchase_timestamp'].dt.day_name()

# 5. Salvar o arquivo unificado
# master.to_csv('OLIST_DATA.csv', index=False)

print("O arquivo 'OLIST_DATA.csv' foi gerado.")
print(f"Total de colunas prontas: {len(master.columns)}")
```
